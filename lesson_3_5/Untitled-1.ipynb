```python
import pandas as pd
vaccine_data = pd.read_csv('vaccinations.csv')
```

```python
vaccine_data_cleaned = vaccine_data[['location', 'date', 'total_vaccinations', 'people_vaccinated_per_hundred']]
```

```python
top_10_vaccinated = vaccine_data_cleaned.sort_values(by='people_vaccinated_per_hundred', ascending=False).drop_duplicates(subset=['location']).head(10)
```

```python
bottom_10_vaccinated = vaccine_data_cleaned.sort_values(by='people_vaccinated_per_hundred', ascending=True).drop_duplicates(subset=['location']).head(10)
```

```python
import matplotlib.pyplot as plt

# 
plt.figure(figsize=(10, 6))
plt.barh(top_10_vaccinated['location'], top_10_vaccinated['people_vaccinated_per_hundred'])
plt.xlabel('Percentage of People Vaccinated')
plt.title('Top 10 Countries with Highest Vaccination Rates')
plt.show()

# 
plt.figure(figsize=(10, 6))
plt.barh(bottom_10_vaccinated['location'], bottom_10_vaccinated['people_vaccinated_per_hundred'])
plt.xlabel('Percentage of People Vaccinated')
plt.title('Top 10 Countries with Lowest Vaccination Rates')
plt.show()
```

```python
print("Top 10 countries with highest vaccination rates:")
print(top_10_vaccinated[['location', 'people_vaccinated_per_hundred']])

print("\nTop 10 countries with lowest vaccination rates:")
print(bottom_10_vaccinated[['location', 'people_vaccinated_per_hundred']])
```
